```python
#A website check password
password = input()
upper = False
digit = False
special = False
special_chars = "!@#$%^&*()-_=+[]{};:'\",.<>?/\\|`~"
for ch in password:
    if ch.isupper():
        upper = True
    if ch.isdigit():
        digit = True
    if ch in special_chars:
        special = True
if upper and digit and special and len(password) >= 8:
    print("Strong Password")
else:
    print("Weak Password")
```

```python
#By using alnum
password=input()
upper=False
digit=False
special=False
for ch in password:
  if ch.isupper():
    upper=True
  if ch.isdigit():
    digit=True
  if ch.isalnum:
    special=True
if upper and digit and special and len(password)>=8:
  print("Strong Password")
else:
  print("Weak Password")
```

```python
#Movie Rating Analyzer
n = int(input())
ratings = []
for i in range(n):
    ratings.append(int(input()))
unique = []
for num in ratings:
    if num not in unique:
        unique.append(num)
print(unique)
for i in range(0,len(unique)):
    for j in range(i+1,len(unique)):
        if(unique[i]>unique[j]):
            unique[i],unique[j]=unique[j],unique[i]
print(unique)
print(unique[-2])
```

```python
#Marathon Runner Tracker
li=[2,3,5,4,6,7,8,2,5,7,8,9,12,13]
count=1
max_count=0
for i in range(1,len(li)):
  if li[i]>li[i-1]:
    count+=1
  else:
    max_count=max(max_count,count)
    count=1
print(max(max_count,count))
```

```python
#Online Voting System
s = "ABACBA"
char_count = {}
for ch in s:
    char_count[ch] = char_count.get(ch, 0) + 1
max_char = max(char_count, key=char_count.get)
print(max_char)
```
