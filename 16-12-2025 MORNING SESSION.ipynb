```python
class Product:
  def __init__(self, id, name,expiry):
    self.id = id
    self.name = name
    self.__expiry = expiry

  def print(self):
    print(self.id, " ", self.name, " ")

  def setExpiry(self, expiry):
    self.__expiry = expiry

  def getExpiry(self):
    return self.__expiry

p = Product(1,"pizza","10-11-2026")
p.print()
print(p.getExpiry())
p.setExpiry("12345")
print(p.getExpiry())
```
