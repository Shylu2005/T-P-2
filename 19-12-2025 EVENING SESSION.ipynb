```python
#Sum of Array Elements
def f_sum(l,a):
    if len(l)==a:
        return 0
    else:
        return l[a]+sum(l,a+1)
l=[1,2,3,4]
print(sum(l))
```

```python
def attack(p):
  if p==0:
    return 0
  else:
    return 1+attack(p-1)
print(attack(5))
```

```python
def read_pages(current, n):
    if current > n:
        return current
    print(f"{current}")
    read_pages(current + 1, n)
n = 10
read_pages(1, n)
```

```python
#HACKER RANK
def catAndMouse(x, y, z):
    dist_a = abs(x - z)
    dist_b = abs(y - z)

    if dist_a < dist_b:
        return "Cat A"
    elif dist_b < dist_a:
        return "Cat B"
    else:
        return "Mouse C"

q = int(input().strip())
for _ in range(q):
    x, y, z = map(int, input().split())
    print(catAndMouse(x, y, z))
```

```python
#HACKER RANK
# Find Maximum element in the list
def find_max(ar,n):
  if n==len(ar):
    return 0
  return max(ar[n],find_max(ar,n+1))

n=int(input())
ar=[]
for i in range(n):
  ar.append(int(input()))
r=find_max(ar, 0)
print("Maximum element in the array is:",r)
```
