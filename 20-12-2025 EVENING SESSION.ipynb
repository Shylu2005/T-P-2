```python
l = list(map(int, input().split()))
flag = True
for i in range(len(l) - 1):
    print(l[i], l[i+1])
    if (l[i] % 2 == l[i+1] % 2):
        flag = False
        break
if flag:
    print("True")
else:
    print("False")
```

```python
l = ['r','g','b','b','g','y','y']
found_odd = False
for elem in l:
    if l.count(elem) % 2 == 1:
        print(elem)
        found_odd = True
        break
if not found_odd:
    print("All are even")
```
