```python
class ChildChota(Exception):
    pass
class BadaException(Exception):
    pass
n=int(input())
try:
    if n<18:
        raise ChildChota("Not Eligible")
    elif n>100:
        raise BadaException("Age id too Big")
    else:
        print("Eligible")
except BadaException as e:
    print(e)
print("Rest of the code")
```

```python
#Online exam system
#A student answers questions one by one
#if the answer is -1 exam ends immediately
#Calculate toytal score until exam ends
#I/P:5
#4 5 3 -1 2
#O/P:12
n = int(input())
answers = list(map(int, input().split()))
total = 0
for ans in answers:
    if ans == -1:
        break
    total += ans
print(total)
```

```python
#without using map()
n = int(input())
answers = []
for _ in range(n):
    answers.append(int(input()))
total = 0
for ans in answers:
    if ans == -1:
        break
    total += ans
print(total)
```

```python
password = input("Enter password: ")
special_characters = "!@#$%^&*()-+"
has_upper = any(ch.isupper() for ch in password)
has_lower = any(ch.islower() for ch in password)
has_digit = any(ch.isdigit() for ch in password)
has_special = any(ch in special_characters for ch in password)
if (len(password) >= 8 and has_upper and has_lower and has_digit and has_special):
    print("Strong password")
else:
    print("Weak password")
```
