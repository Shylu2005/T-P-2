```python
#Student Marks Analyzer
students = [
    {"name": "ravi", "marks": 65},
    {"name": "anu", "marks": 85},
    {"name": "kiran", "marks": 75},
    {"name": "meena", "marks": 88}
]
total_marks = sum(student["marks"] for student in students)
class_average = total_marks / len(students)
above_average_students = [student["name"] for student in students if student["marks"] > class_average]
print("Students who scored above class average:")
for name in above_average_students:
    print(name)
```

```python
#Hallow Square
for i in range(1,6):
    for j in range(1,6):
        if i==1 or i==5 or j==1 or j==5:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

```python
n=int(input())
while n>= 10:
    sum=0
    while n>0:
        digit=n%10
        sum=sum+digit
        n=n//10
    n=sum
print(sum)
```

```python
#Factorial of a number using recursion
def f(n):
  if n==1 or n==0:
    return 1
  else:
    return n*f(n-1)
n=int(input())
print(f(n))
```

```python
#power of a number using recursion
def f(n,m):
  if m==0:
    return 1
  else:
    return n*f(n,m-1)
n=int(input())
m=int(input())
print(f(n,m))
```
