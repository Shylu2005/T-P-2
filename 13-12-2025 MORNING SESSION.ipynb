```python
#check how many special characters, alphabets,lowercase,uppercase,digits present inside a string
s1="python programming @Lang)(12"
upper=lower=digit=special=0
for ch in s1:
  if ch.islower():
    lower+=1
  elif ch.isupper():
    upper+=1
  elif ch.isdigit():
    digit+=1
  else:
    special+=1
print("upper case",upper)
print("lower case",lower)
print("digit case",digit)
print("special case",special)
```

```python
#Reverse each word in a string
s1="HI python"
word= ""
for c in s1:
  if(c!=' '):
    word = c + word
  else:
    print(word,end=" ")
    word = ""
print(word)
```

```python
#write a program to remove duplicates in the current string
s1="aabcwweq"
result=""
for c in s1:
  if c not in result:
    result=result+c
print(result)
```

```python
a=input("Enter a First string:")
b=input("Enter a Second string:")
print(sorted(a))
print(sorted(b))
if len(a)!=len(b):
  print("Not an Anagram")
else:
  if sorted(a)==sorted(b):
    print("Anagram")
  else:
    print("Not Anagram")
```
