```python
class Bank:
  def __init__(self,name,balance):
    self.name=name
    self.__balance=balance
  def deposit(self,balance):
    if(self.__balance>0):
      self.__balance+=balance
      print("Deposite success")
    else:
      print("Not success")
  def withDraw(self,withDraw_money):
    if(self.__balance >0):
      self.__balance=self.__balance-withDraw_money
      return self.__balance
    else:
      print("invalid amount")
b1=Bank("sbi",100)
b1.deposit(1000)
print(b1.withDraw(100))
```

```python
#method1:without parameter without return type
#method2:with parameter without return type
#method3:without parameter with return type
#method4:with parameter with return type
class Cal:
    def method1(self):
        print("without paramter without return type")
    def method2(self,a,b):
        print(a,b)
    def method3(self):
        a=10
        b=20
        if a<b:
            return "yes a is less than b"
        else:
            return "a is greater than b"
    def method4(self,firstName,lastName):
        return firstName,lastName


c=Cal()
c.method1()
c.method2(10,20)
result=c.method3()
print(result)
print(c.method4("sr","university"))
```

```python
#Multilevel Inheritance
class A:
    def displayA(self):
        print("This is class A")
class B(A):
    def displayB(self):
        print("This is class B")
class C(B):
    def displayC(self):
        print("This is class C")
obj = C()
obj.displayA()
obj.displayB()
obj.displayC()
```

```python
#Hierarchical Inheritance - one parent class extending by multiple child class
class Parent:
    def property(self):
        print("Parent property")
class Child(Parent):
    pass
class Child2(Parent):
    pass
c2=Child2()
c2.property()
```

```python
#polymorphysm
class parent:
  def marry(self):
    print("Marry Indian Girl")
class child(parent):
  def marry(self):
    print("No marriage")
c=child()
c.marry()

class A:
  def __init__(self,num1, num2):

    self.num1 = num1
    self.num2 = num2

class B(A): # Un-nested class B
  def __init__(self, num1, num2, num3, num4):
    super().__init__(num1, num2) # Correctly calls A's __init__
    self.num3 = num3
    self.num4 = num4
  def printData(self):
    print(self.num1, self.num2,self.num3, self.num4)
b=B(10, 20, 30, 40) # Instantiate B outside of class A
b.printData()
```

```python
#Abstraction
from abc import ABC, abstractmethod
class Bank(ABC):
  @abstractmethod
  def rateOfInterest(self):
    pass
class Sbi(Bank):
  def rateOfInterest(self):
    print("7% of Rate of interest")

  @staticmethod
  def new():
    print("Sbi Bank")
s=Sbi()
s.rateOfInterest()
Sbi.new()
```
