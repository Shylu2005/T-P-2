```python
#afternoon session
#linear search
l=[8,9,1,10,3,2,1]
find=10
for i in range(0,len(l)):
  if l[i]==find:
    print("Element fount at index:",i)
    break
  else:
    print("Element not found")
```

```python
#by using index
def linear_search(arr, find):
    for i in range(len(arr)):
        if arr[i] == find:
            print("Element found at index", i)
            return
    print(-1)

arr = [5, 10, 15, 20, 25]
linear_search(arr, 100)
linear_search(arr,10)
linear_search(arr,15)
```

```python
#binary search
def BinarySearch(l,find):
    low=0
    high=len(l)-1
    while low<=high:
        mid=(low+high)//2
        if l[mid]==find:
            return mid
        elif l[mid]<find:
            low=mid+1
        else:
            high=mid-1
    return -1
l=[3,4,5,7,8,19,23]
print(BinarySearch(l,19))
```
