```python
#double your number and then find even numbers and store in list
l=[1,2,3,4,5]
double = list(map(lambda x:x*2,l))
print(double)
even = list(filter(lambda x:x%2==0,double))
print(even)
```

```python
#Another method - double your number and then find even numbers and store in list
l=[1,2,3,4,5]
double = list(map(lambda x:x*2,l))
even = list(filter(lambda x:x%2==0,double))
print(even)
```
