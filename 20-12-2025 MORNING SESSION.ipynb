```python
#count pairs whose sum is divisible by 2
l=list(map(int,input().split()))
count=0
for i in range(0,len(l)):
  for j in range(i+1,len(l)):
    if (l[i]+l[j])%2==0:
      count+=1
print(count)
```

```python
#print max count number
a=[7,4,8,2,9]
max_count=0
count=0
for i in a:
  if(i>max_count):
    count+=1
    max_count=i
print(count)
```

```python
#print maximum guests
entering = [5,7,0,5,0]
leaving  = [1,2,1,2,5]
t=int(input())
max_guests=0
guest=0
for i in range(0,t):
    if max_guests>=0:

        guest+=entering[i]-leaving[i]
        max_guests=max(max_guests,guest)
print(max_guests)
```

```python
num=int(input("Enter a number:"))
digit_sum=0
temp=num
while temp>0:
  digit_sum += temp%10
  temp//=10
print("Sum of digits:",digit_sum)
if(num%digit_sum==0):
  print("Good Number")
else:
  print("Bad Number")
```
