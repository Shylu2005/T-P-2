```python
#nested for loops
n=int(input())
for i in range(n):
  for j in range(i,-1,-1):
    print(j+1,end=" ")
  print()
```

```python
n=int(input())
for i in range(n):
  print("i",i)
  for j in range(i,-1,-1):
    print("J:",j)
    print(j+1,end=" ")
    print("j+1",j)
  print()
```

```python
n=6
for i in range(n):
  print("Before enter i:",i)
  for j in range(i+1):
    print("j in nested loop:",j)
    print(j+1,end=" ")
  print("out of nested loop")
```

```python
rows=int(input("enter rows size for the pattern:"))
for i in range(1,rows+1):
  for j in range(rows - i):
    print(" ",end=" ")
  for k in range(1,2*i):
    print("*",end=" ")
  print()
```

```python
rows=int(input("enter rows size for the pattern:"))
for i in range(rows,0,-1):
  for j in range(rows - i):
    print(" ",end=" ")
  for k in range(1,2*i):
    print("*",end=" ")
  print()
```

```python
#Upper part
rows=int(input("enter rows size for the pattern:"))
for i in range(1,rows+1):
  for j in range(rows - i):
    print(" ",end=" ")
  for k in range(1,2*i):
    print("*",end=" ")
  print()
#Lower part
for i in range(rows-1,0,-1):
  for j in range(rows - i):
    print(" ",end=" ")
  for k in range(1,2*i):
    print("*",end=" ")
  print()
```
