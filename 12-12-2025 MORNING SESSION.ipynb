```python
#write a program to print factorial of a number
#for loop
n=int(input())
fact=1
for i in range(1, n+1):
  fact = fact * i
print(fact)
```

```python
#write a program to print factorial of a number
#while loop
n=int(input())
i=1
fact=1
while i<=n:
  fact = fact * i
  i+=1
print(fact)
```
